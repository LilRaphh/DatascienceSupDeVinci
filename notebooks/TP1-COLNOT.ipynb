## Objectif
Vérifier si les calls API de la plateforme FastAPI

## Structure
- Récupération du dataset
- Endpoints de nettoyage (fit/transform)
- Rapport qualité des données


```python
import requests
import pandas as pd


BASE_URL = "http://api:8000"
# Générer le dataset
response = requests.post(
    f"{BASE_URL}/dataset/generate",
    json={
        "phase": "clean",
        "seed": 42,
        "n": 1000
    }
)

data = response.json()
dataset_id = data["meta"]["dataset_id"]

print(f"Dataset ID: {dataset_id}")
print(f"Nombre de lignes: {data['result']['n_rows']}")
print(f"Colonnes: {data['result']['columns']}")

# Afficher un échantillon
df_sample = pd.DataFrame(data['result']['data_sample'])
df_sample.head(10)
```
